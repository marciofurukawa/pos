### Consulta 2

-- Quantidade de postagens dos alunos no forum nos últimos 30 dias:

```select u.id_user, sum(f.id_forum is not null) postagens 
     from usuario u 
     LEFT OUTER JOIN (select id_user, id_forum 
                        from forum 
                       where data_post>strftime('%s','now') - 606024*30) AS f 
     ON u.id_user = f.id_user 
    group by u.id_user 
    order by postagens asc;
```

````pyspark
# cria dataframe
df_usuarios = spark.read.csv(
    "wasbs://dadoscru@storhack.blob.core.windows.net/usuarios.csv",
    header=True,
    sep=",",
)

df_usuarios.createTempView("usuarios")

df_forum = spark.read.csv(
    "wasbs://dadoscru@storhack.blob.core.windows.net/AVA.forum.csv",
    header=True,
    sep=",",
)

df_forum.createTempView("forum")
````

````pyspark
spark.sql("""
    select 
        data_post,
        from_unixtime(data_post) as data_post_date,
        to_timestamp(from_unixtime(data_post)) as data_post_ts,
        to_date(from_unixtime(data_post), "yyyy-MM-dd hh24:mi:ss") as a,
        current_timestamp(),
        to_date(to_timestamp(from_unixtime(data_post)), "yyyy-MM-dd HH:mm:ss") as b
    from forum
    limit 10
""").show(10, -1)
````

````pyspark
spark.sql("""
    select 
        data_post,
        from_unixtime(data_post) as data_post_date,
        to_timestamp(from_unixtime(data_post)) as data_post_ts,
        to_date(from_unixtime(data_post), "yyyy-MM-dd hh24:mi:ss") as a,
        current_timestamp(),
        to_date(current_timestamp(), "yyyy-MM-dd HH:mm:ss.SSS") as b
    from forum
    limit 10
""").printSchema()
````

````pyspark
spark.sql("""
    select 
        u.id_user, 
        sum(f.id_forum is not null) postagens 
    from usuarios u 
    LEFT OUTER JOIN (
                    select 
                         id_user, 
                         id_forum 
                    from forum 
                    where from_unixtimestamp(data_post) > (current_timestamp() - 606024*30) AS f 
     ON u.id_user = f.id_user 
    group by u.id_user 
    order by postagens asc
""").show(10, -1)
````
